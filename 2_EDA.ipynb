```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
df.sample(4)
```

```python
df['Survived'].value_counts()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x = df['Survived'].value_counts().index, y = df['Survived'].value_counts())
plt.title('Value Counts of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Tasa de Supervivencia por Clase
plt.figure(figsize=(10, 6))

# Crear tabla cruzada de Pclass y Survived
survival_by_class = pd.crosstab(df['Pclass'], df['Survived'])

# Crear grfico de barras agrupadas
survival_by_class.plot(kind='bar', 
                        color=['#440154', '#31688e'],
                        figsize=(10, 6),
                        width=0.8)
                    

plt.title('Tasa de Supervivencia por Clase', fontsize=14, pad=20)
plt.xlabel('Clase', fontsize=12)
plt.ylabel('Total', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12, 8))

# colores para cada una de las clases
color = ['#240750', '#344C64']

sns.countplot(df, x = 'Embarked', hue = 'Survived', palette=color) 
plt.legend(title = 'Supervivencia', labels = ['No sobrevivi', 'S sobrevivi'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))

sns.histplot(df, x = 'Age', kde = True)
plt.title('Distribucin de Edades')
plt.axvline(df['Age'].mean(), color = 'red')
plt.axvline(df['Age'].median(), color = 'green')
plt.axvline(df['Age'].mode()[0], color = 'black')
plt.show()

# Print the results
print(f"Media: {df['Age'].mean()}")
print(f"Mediana: {df['Age'].median()}")
print(f"Moda: {df['Age'].mode()[0]}")
```

```python
plt.figure(figsize = (12, 8))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), 
            annot=True, 
            cbar=True,  # Activar la barra de colores
            cmap='coolwarm',  # Cambiar la paleta de colores
            cbar_kws={'label': 'Nivel de correlacin'})  # Etiqueta para la barra de colores
plt.title('Correlacin de variables numricas')
plt.show()
```

```python
# 1. Distribucin de edades de los pasajeros
plt.figure(figsize=(12, 8))
sns.histplot(df['Age'], bins=30, kde=True, color='#440154')
plt.title('Distribucin de Edades de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# 2. Cantidad de sobrevivientes por clase de pasajero (Pclass)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Pclass', hue='Survived', palette=['#440154', '#31688e'])
plt.title('Cantidad de Sobrevivientes por Clase de Pasajero')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.show()
```

```python
# 3. Cantidad de sobrevivientes por gnero (Sex)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Sex', hue='Survived', palette=['#440154', '#31688e'])
plt.title('Cantidad de Sobrevivientes por Gnero')
plt.xlabel('Gnero')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.show()
```

```python
# 4. Cantidad de sobrevivientes por puerto de embarque (Embarked)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Embarked', hue='Survived', palette=['#440154', '#31688e'])
plt.title('Cantidad de Sobrevivientes por Puerto de Embarque')
plt.xlabel('Puerto de Embarque')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.show()
```

```python
# 5. Agrupar edades en categoras y mostrar tasa de supervivencia
# Definir categoras de edad
bins = [0, 2, 12, 18, 25, 60, 100]
labels = ['Infante', 'Nio', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Crear grfica de supervivencia por grupo de edad
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Age_Group', hue='Survived', palette=['#440154', '#31688e'])
plt.title('Tasa de Supervivencia por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6. Conteo de pasajeros en cada clase (Pclass)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Pclass', palette='viridis')
plt.title('Conteo de Pasajeros por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
# 7. Distribucin de gnero dentro de cada clase de pasajero
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Pclass', hue='Sex', palette=['#440154', '#31688e'])
plt.title('Distribucin de Gnero por Clase de Pasajero')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.legend(title='Gnero')
plt.show()
```

```python
# 8. Distribucin de clase de pasajero por puerto de embarque
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Embarked', hue='Pclass', palette='viridis')
plt.title('Distribucin de Clase por Puerto de Embarque')
plt.xlabel('Puerto de Embarque')
plt.ylabel('Cantidad')
plt.legend(title='Clase')
plt.show()
```

```python
# 9. Conteo de pasajeros por gnero (Sex)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Sex', palette=['#440154', '#31688e'])
plt.title('Conteo de Pasajeros por Gnero')
plt.xlabel('Gnero')
plt.ylabel('Cantidad')
plt.show()
```

```python
# 10. Valores mximo y mnimo de la columna Age
print(f"Edad mxima: {df['Age'].max()}")
print(f"Edad mnima: {df['Age'].min()}")
```

```python
# 11. Distribucin de edades con lneas de media, mediana y moda
plt.figure(figsize=(12, 8))
sns.histplot(df['Age'], bins=30, kde=True, color='#440154')
plt.axvline(df['Age'].mean(), color='red', linestyle='--', linewidth=2, label=f"Media: {df['Age'].mean():.2f}")
plt.axvline(df['Age'].median(), color='green', linestyle='--', linewidth=2, label=f"Mediana: {df['Age'].median():.2f}")
plt.axvline(df['Age'].mode()[0], color='blue', linestyle='--', linewidth=2, label=f"Moda: {df['Age'].mode()[0]:.2f}")
plt.title('Distribucin de Edades con Estadsticas')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
# 12. Diagrama de caja (boxplot) con outliers en la columna Age
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['Age'], color='#440154')
plt.title('Boxplot de Edad - Deteccin de Outliers')
plt.xlabel('Edad')
plt.show()
```

```python
# 13. Valores mximo y mnimo de la columna Fare
print(f"Tarifa mxima: {df['Fare'].max()}")
print(f"Tarifa mnima: {df['Fare'].min()}")
```

```python
# 14. Distribucin de tarifas con lneas de media, mediana y moda
plt.figure(figsize=(12, 8))
sns.histplot(df['Fare'], bins=30, kde=True, color='#31688e')
plt.axvline(df['Fare'].mean(), color='red', linestyle='--', linewidth=2, label=f"Media: {df['Fare'].mean():.2f}")
plt.axvline(df['Fare'].median(), color='green', linestyle='--', linewidth=2, label=f"Mediana: {df['Fare'].median():.2f}")
plt.axvline(df['Fare'].mode()[0], color='blue', linestyle='--', linewidth=2, label=f"Moda: {df['Fare'].mode()[0]:.2f}")
plt.title('Distribucin de Tarifas con Estadsticas')
plt.xlabel('Tarifa')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
# 15. Diagrama de caja (boxplot) con outliers en la columna Fare
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['Fare'], color='#31688e')
plt.title('Boxplot de Tarifa - Deteccin de Outliers')
plt.xlabel('Tarifa')
plt.show()
```

```python
# 16. Filtrar y mostrar pasajeros que pagaron 0 en tarifa
pasajeros_tarifa_cero = df[df['Fare'] == 0]
print(f"Nmero de pasajeros con tarifa 0: {len(pasajeros_tarifa_cero)}")
pasajeros_tarifa_cero
```

```python
# 17. Conteo de pasajeros por nmero de hermanos/esposos a bordo (SibSp)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='SibSp', palette='viridis')
plt.title('Conteo de Pasajeros por Nmero de Hermanos/Esposos a Bordo (SibSp)')
plt.xlabel('Nmero de Hermanos/Esposos')
plt.ylabel('Cantidad')
plt.show()
```

```python
# 18. Tasa de supervivencia basada en el nmero de hermanos/esposos a bordo (SibSp)
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='SibSp', hue='Survived', palette=['#440154', '#31688e'])
plt.title('Tasa de Supervivencia por Nmero de Hermanos/Esposos a Bordo (SibSp)')
plt.xlabel('Nmero de Hermanos/Esposos')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.show()
```

```python
# 19. Porcentaje de supervivencia para cada valor de SibSp
survival_by_sibsp = df.groupby('SibSp')['Survived'].agg(['sum', 'count'])
survival_by_sibsp['percentage'] = (survival_by_sibsp['sum'] / survival_by_sibsp['count'] * 100).round(2)
survival_by_sibsp.columns = ['Sobrevivieron', 'Total', 'Porcentaje_Supervivencia']
print("Porcentaje de Supervivencia por SibSp:")
print(survival_by_sibsp)
```

```python
# 20. Mapa de calor (heatmap) con matriz de correlacin de caractersticas numricas
plt.figure(figsize=(12, 8))
numeric_columns = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            linewidths=1, cbar_kws={'label': 'Correlacin'})
plt.title('Matriz de Correlacin de Caractersticas Numricas')
plt.tight_layout()
plt.show()
```
